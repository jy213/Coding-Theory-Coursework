```julia
# Q9.2.a
# Q9_2_data_r :: Int64
# Q9_2_data_q :: Int64
# Q9_2_a_ans_H :: Union{Missing, Matrix{Int64}}

function generate_vec(r, q)
    non_zero_vec = []
    for i in 1:(q^r - 1)
        vec = digits(i, base=q, pad=r)
        if sum(vec) != 0
            push!(non_zero_vec, vec)
        end
    end
    return non_zero_vec  
end

function sort_H(vec)
    sum = 0
    for i in 1:length(vec)
        sum += vec[i] * 7^(i - 1)    
    end
    return sum
end 

function find_H(r, q)

    non_zero_vec = generate_vec(r, q)

    H = []
    for vec in non_zero_vec
        # Check if the vector is a scalar multiple of any vector already in H
        is_multiple = false
        for factor in 1:(q-1)
            if mod.(vec * factor, q) in H
                is_multiple = true
                break
            end
        end

        if !is_multiple
            push!(H, vec)
        end
    end

    H = hcat(H...)
    return sortslices(H, dims=2, by=sort_H)
    # return H
end

r = CW3_DATA.Q9_2_data_r
q = CW3_DATA.Q9_2_data_q

CW3_DATA.Q9_2_a_ans_H = find_H(r, q)
```

```julia
# Q9.2.c
# Q9_2_c_ans_is_perfect_code :: Union{Missing,Bool}

function is_perfect(r, q, n)
    t = 1
    hamming_bound = 0

    for i in 0:t
        hamming_bound += binomial(n, i) * (q - 1)^i
    end

    if q^r == hamming_bound
        return true
    else
        return false
    end
end

r = CW3_DATA.Q9_2_data_r
q = CW3_DATA.Q9_2_data_q
n = CW3_DATA.Q9_2_b_ans_n
CW3_DATA.Q9_2_c_ans_is_perfect_code = is_perfect(r, q, n)
```

```julia
# Q9.3.a

using Primes
function is_primitive_root(p)
    
    phi = p - 1
    for a in 2:phi
        is_primitive = true
        for i in keys(Primes.factor(phi))
            order = div(phi, i)
            if powermod(a, order, p) == 1
                is_primitive = false
                break
            end
        end
        if is_primitive
            return a
        end
    end
end

p = CW3_DATA.Q9_3_data_p
CW3_DATA.Q9_3_a_ans_alpha = is_primitive_root(p)
```

```julia
# Q9.3.b
# Q9_3_data_n::Int
# Q9_3_data_k1::Int
# Q9_3_data_k2::Int
# Q9_3_data_p::Int


function RS_generator(n, k, p, alpha)
    G = zeros(Int64, k, n)
    for i in 1:k
        for j in 1:n
            G[i, j] = powermod(alpha, (i-1)*(j-1), p)
        end
    end
    
    H = zeros(Int64, n-k, n)
    for i in 1:(n-k)
        for j in 1:n
            H[i, j] = powermod(alpha, i*(j-1), p)
        end
    end
    return G, H
end

n = CW3_DATA.Q9_3_data_n
k1 = CW3_DATA.Q9_3_data_k1
k2 = CW3_DATA.Q9_3_data_k2
CW3_DATA.Q9_3_b_ans_generator, CW3_DATA.Q9_3_b_ans_parity = RS_generator(n, k1, p, CW3_DATA.Q9_3_a_ans_alpha)
display(CW3_DATA.Q9_3_b_ans_generator)
display(CW3_DATA.Q9_3_b_ans_parity)
```

```julia
function det_cofactor_FF(M, R)
  
    determinant = R(0)
    dim = size(M, 2)

    if dim == 2
        determinant = M[1, 1] * M[2, 2] - M[1, 2] * M[2, 1]
        return determinant
    end
    
    for idx in 1:dim
        M_minor = zeros(R, dim-1, dim-1)
        M_minor_new = zeros(R, dim-1, dim+2)
        M_new = zeros(R, dim-1, dim+3)
        M_new[1:(dim-1), 2:dim+1] = M[2:dim, 1:dim]
        M_minor_new[1:(dim-1), 1:idx] = M_new[1:(dim-1), 1:idx]
        M_minor_new[1:(dim-1), (idx+1):(dim+2)] = M_new[1:(dim-1), (idx+2):(dim+3)]
        M_minor[1:(dim-1), 1:(dim-1)] = M_minor_new[1:(dim-1), 2:(dim)]
  
        cofactor = (-1)^(1+idx) * det_cofactor_FF(M_minor, R)
      
        determinant += (M[1, idx] * cofactor)
    end
    
    return determinant
end

function inv_cofactor_FF(M, m, R, det_M)
    dim = size(M, 2)
    M_c = zeros(R, dim, dim)

    for i in 1:dim
        for j in 1:dim
            M_minor = zeros(R, dim-1, dim-1)
            M_minor_new = zeros(R, dim+2, dim+2)
            M_new = zeros(R, dim+3, dim+3)
            M_new[2:dim+1, 2:dim+1] = M[1:dim, 1:dim]
            M_minor_new[1:i, 1:j] = M_new[1:i, 1:j]
            M_minor_new[1:i, (j+1):(dim+2)] = M_new[1:i, (j+2):(dim+3)]
            M_minor_new[(i+1):(dim+2), 1:j] = M_new[(i+2):(dim+3), 1:j]
            M_minor_new[(i+1):(dim+2), (j+1):(dim+2)] = M_new[(i+2):(dim+3), (j+2):(dim+3)]
            M_minor[1:(dim-1), 1:(dim-1)] = M_minor_new[2:dim, 2:dim]
            M_c[i, j] = (-1)^(i+j) * det_cofactor_FF(M_minor, R)
        end
    end

    M_adj = transpose(M_c)
    M_inv = invmod(det_M, m) .* M_adj
    return mod.(M_inv, m)
end

function Find_message(a, e, y, G)
    F_base = GF(2)
    # Create PolynomialRing based on this field
    Rz = PolynomialRing(F_base, "z")[1]
    Rx, gen = PolynomialRing(Rz, "x")
    # Create the irreducible polynomial based on coefficients vector
    f = Rz(a)
    # Create the extension finite field with q = base^deg
    F_32 = FiniteField(f, "z")
    # Get the primitive element
    m_y, n_y = size(y)
    y_F = zeros(Rz, n_y)
    for j in 1:n_y
        y_F[j] = Rz(y[1, j])
    end
    c = mod.(y_F-e, f)
    c_vec = Matrix{Vector{Int64}}(undef, 1, length(c))
    for i in 1:length(c)
        c_vec[1, i] = [Int(lift(m)) for m in coefficients(c[i])] != [] ? [Int(lift(m)) for m in coefficients(c[i])] : [0]
    end
    k, n = size(G)
    m_vec = Matrix{Vector{Int64}}(undef, 1, k)
    m_vec[1, 1:k] = c_vec[1, 1:k]
    # det = det_cofactor_FF(G_F, Rz)
    # G_inv_F = inv_cofactor_FF(G, f, Rz, det)
    # G_new = G[1:21,1:21]
    # G_1 = MatrixSpace(F, 21, 21)()
    # G_inv_F = MatrixSpace(F, 21, 21)()

    # for i in 1:21
    #     for j in 1:21
    #         G_1[i, j] = vector_to_poly(G_new[i, j], z)
    #     end
    # end
    # G_inv_F = inv(G_1)
    # m = mod.(c*G_inv_F)
    return m_vec
end
CW3_DATA.Q10_5_ans_message = Find_message(CW3_DATA.Q10_data_polynomial, e_F, CW3_DATA.Q10_2_data_y, G_F)
```
